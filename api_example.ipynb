## Example API Call
This is an example notebook that calls an API and puts the response into a dataframe for additional processing.

```python
import requests
import pandas as pd
from getpass import getpass
```

#### Call the API and Format Response

```python
# prompt for the api key as it might be a secret
api_key = getpass()
```

```python
# format the API by appending api_key to the end of the url
api = f"https://feeds.datagolf.com/preds/in-play?tour=pga&dead_heat=yes&odds_format=percent&file_format=json&key={api_key}"
```

```python
# call the API and save the response as variable `response`
response = requests.get(url=api)
```

```python
# the API returns json (so response.json) then all the data is in the `data` key of the json response
data = pd.DataFrame.from_records(response.json()['data'])
```
